### Configuración de DataFrames

En esta sección se realizan las transformaciones iniciales necesarias para el análisis: formateo de fechas, extracción de componentes de tiempo (hora, día, mes, año) y estandarización de columnas clave.

```python
pd.set_option('display.max_rows', None)

FECHA_ACTUAL_DT = datetime.now()
NUMERO_DIA = FECHA_ACTUAL_DT.day
MES_ACTUAL = FECHA_ACTUAL_DT.month
ANIO_ACTUAL = FECHA_ACTUAL_DT.year
FECHA_HOY_DATE = datetime.now().date()
hoy = pd.Timestamp.today()

print(f"Fecha de proceso: {FECHA_ACTUAL_DT}")

# ==========================================
# 2. TRANSFORMACIÓN: PQRSD SERVICIOS
# ==========================================
if not df_pqrsd_servicios.empty:
    df_pqrsd_servicios['Fecha_Cierre_PQRS'] = pd.to_datetime(df_pqrsd_servicios['Fecha_Cierre_PQRS'])
    df_pqrsd_servicios['Fecha_Creacion'] = pd.to_datetime(df_pqrsd_servicios['Fecha_Creacion'])
    df_pqrsd_servicios['fecha_creacion_agenda'] = pd.to_datetime(df_pqrsd_servicios['fecha_creacion_agenda'])

    df_pqrsd_servicios['Hora_Cierre'] = df_pqrsd_servicios['Fecha_Cierre_PQRS'].dt.hour.fillna(-1).astype(int)
    df_pqrsd_servicios['dia_fecha_Cierre'] = df_pqrsd_servicios['Fecha_Cierre_PQRS'].dt.day.fillna(-1).astype(int)
    df_pqrsd_servicios['mes_fecha_Cierre'] = df_pqrsd_servicios['Fecha_Cierre_PQRS'].dt.month.fillna(-1).astype(int)
    df_pqrsd_servicios['año_fecha_Cierre'] = df_pqrsd_servicios['Fecha_Cierre_PQRS'].dt.year.fillna(-1).astype(int)

    df_pqrsd_servicios['Hora_Creacion'] = df_pqrsd_servicios['Fecha_Creacion'].dt.hour.fillna(-1).astype(int)
    df_pqrsd_servicios['dia_Fecha_Creacion'] = df_pqrsd_servicios['Fecha_Creacion'].dt.day.fillna(-1).astype(int)
    df_pqrsd_servicios['mes_Fecha_Creacion'] = df_pqrsd_servicios['Fecha_Creacion'].dt.month.fillna(-1).astype(int)
    df_pqrsd_servicios['año_Fecha_Creacion'] = df_pqrsd_servicios['Fecha_Creacion'].dt.year.fillna(-1).astype(int)


# ==========================================
# 3. TRANSFORMACIÓN: PARADAS (PR)
# ==========================================
def extraer_horas(tiempo):
    """Función auxiliar para extraer hora de string 'HH:MM'"""
    if isinstance(tiempo, str):
        partes = tiempo.split(':')
        return int(partes[0])
    return 0

if not df_paradas.empty:
    df_paradas['Hora_total_parada'] = df_paradas['tiempo_parada'].apply(extraer_horas)

    cols_fecha_paradas = ['fecha_ceracion_pqrs', 'fecha_cierre_ticket', 'dateparada', 'finishparada']
    for col in cols_fecha_paradas:
        df_paradas[col] = pd.to_datetime(df_paradas[col])

    df_paradas['Dia_Parada'] = df_paradas['dateparada'].dt.day
    df_paradas['dia_semana'] = df_paradas['dateparada'].dt.dayofweek
    df_paradas['Hora_inicio'] = df_paradas['dateparada'].dt.hour
    df_paradas['Hora_fin'] = df_paradas['finishparada'].dt.hour


# ==========================================
# 4. PREPARACIÓN DE DATAFRAMES BASE (FILTROS INICIALES)
# ==========================================

if not df_pqrsd_servicios.empty:
    df_base_pqrs = df_pqrsd_servicios.copy()
    df_base_pqrs = df_base_pqrs[['ID','ID_Beneficiario', 'CATEGORIA', 'SUBCATEGORIA', 'Estado','Fecha_Creacion']]
    df_base_pqrs = df_base_pqrs.rename(columns={'ID_Beneficiario': 'ID_Beneficiario_PQRSD', 'Estado':'Estado_PQRSD'})

    condicion_cat = df_base_pqrs['CATEGORIA'] == 'CD-CENTROS DE GESTIÓN SERVICIO'
    condicion_estado = df_base_pqrs['Estado_PQRSD'] != 'Cerrar'
    df_base_pqrs = df_base_pqrs[condicion_cat & condicion_estado]
else:
    df_base_pqrs = pd.DataFrame()


if not df_sitios_activos.empty:
    df_base_sitios = df_sitios_activos.copy()
    df_base_sitios = df_base_sitios[['Id_Beneficiario','Estado_Sitio']]
    df_base_sitios = df_base_sitios.rename(columns={'Id_Beneficiario': 'ID_Beneficiario_SITIOS'})
    df_base_sitios.columns = df_base_sitios.columns.str.strip().str.upper()
else:
    df_base_sitios = pd.DataFrame()


if not df_contadores.empty:
    df_contadores = df_contadores[['IDENTIFICADOR BENEFICIARIO', 'DISPOSITIVO', 'MAC', 'CONTADOR DEL DÍA','TIPO DE ALERTA']]
    df_contadores = df_contadores.rename(columns={'IDENTIFICADOR BENEFICIARIO': 'IDENTIFICADOR BENEFICIARIO CONTADOR'})

if not df_contadores_velocidad.empty:
    df_contadores_velocidad = df_contadores_velocidad[['TELEPUERTO', 'ZONA', 'DISPOSITIVO','DIAS SIN MEDICION']]
    df_contadores_velocidad['DISPOSITIVO'] = df_contadores_velocidad['DISPOSITIVO'].astype(str).str.split('_').str[0]

#voy aqui
upper_cols = lambda df: df.rename(columns=lambda x: x.strip().upper())

df_falla_ext = df_base_pqrs[df_base_pqrs['SUBCATEGORIA'] == 'CD-FALLA SIN TRÁFICO DE INTERNET 24 HORAS AP EXTERIOR'].copy()
df_falla_int = df_base_pqrs[df_base_pqrs['SUBCATEGORIA'] == 'CD-FALLA SIN TRÁFICO DE INTERNET 24 HORAS AP INTERIOR'].copy()
df_ext_24h   = df_base_pqrs[df_base_pqrs['SUBCATEGORIA'] == 'CD-SIN TRÁFICO DE INTERNET 24 HORAS AP EXTERIOR'].copy()
df_int_24h   = df_base_pqrs[df_base_pqrs['SUBCATEGORIA'] == 'CD-SIN TRÁFICO DE INTERNET 24 HORAS AP INTERIOR'].copy()
df_vel_5_dias = df_base_pqrs[df_base_pqrs['SUBCATEGORIA'] == 'CD-FALLA EN EJECUCIÓN DE PRUEBA DE VELOCIDAD 5 DÍAS CALENDARIO'].copy()
df_vel_8_dias = df_base_pqrs[df_base_pqrs['SUBCATEGORIA'] == 'CD-MEDICIÓN DIRECTA DE VELOCIDAD EFECTIVA DE TRANSMISIÓN DE DATOS'].copy()

lista_dfs_pqrs = [df_falla_ext, df_falla_int, df_ext_24h, df_int_24h, df_vel_5_dias, df_vel_8_dias]
for i, df in enumerate(lista_dfs_pqrs):
    df.columns = df.columns.str.strip().str.upper()

if not df_contadores.empty:
    #df_cont_outdoor = df_contadores[df_contadores['DISPOSITIVO'].str.contains('_Outdoor', na=False)].copy()
    condicion_mikrotik_outdoor = df_contadores['DISPOSITIVO'].str.contains('_Outdoor', na=False)
    condicion_fortinet_outdoor = df_contadores['TIPO DE ALERTA'].str.contains('outdoor', na=False)
    df_cont_outdoor = df_contadores[condicion_mikrotik_outdoor | condicion_fortinet_outdoor].copy().reset_index(drop=True)

    condicion_mikrotik_indoor = df_contadores['DISPOSITIVO'].str.contains('_Indoor', na=False)
    condicion_fortinet_indoor = df_contadores['TIPO DE ALERTA'].str.contains('indoor', na=False)
    df_cont_indoor = df_contadores[condicion_mikrotik_indoor | condicion_fortinet_indoor].copy().reset_index(drop=True)


    df_cont_out_8_dias = df_cont_outdoor[df_cont_outdoor['CONTADOR DEL DÍA'] >= 8].copy()
    df_cont_in_8_dias  = df_cont_indoor[df_cont_indoor['CONTADOR DEL DÍA'] >= 8].copy()

    df_cont_out_1_dias = df_cont_outdoor[df_cont_outdoor['CONTADOR DEL DÍA'] >= 1].copy()
    df_cont_in_1_dias  = df_cont_indoor[df_cont_indoor['CONTADOR DEL DÍA'] >= 1].copy()

    lista_dfs_cont = [df_cont_out_8_dias, df_cont_in_8_dias, df_cont_out_1_dias, df_cont_in_1_dias]
    for df in lista_dfs_cont:
        df.columns = df.columns.str.strip().str.upper()
else:
    df_cont_out_8_dias = pd.DataFrame()
    df_cont_in_8_dias = pd.DataFrame()
    df_cont_out_1_dias = pd.DataFrame()
    df_cont_in_1_dias = pd.DataFrame()


if not df_contadores_velocidad.empty:
    df_cont_vel_5_dias = df_contadores_velocidad[df_contadores_velocidad['DIAS SIN MEDICION'] >= 5].copy()
    df_cont_vel_8_dias  = df_contadores_velocidad[df_contadores_velocidad['DIAS SIN MEDICION'] >= 8].copy()

    lista_dfs_cont_vel = [df_cont_vel_5_dias, df_cont_vel_8_dias]
    for df in lista_dfs_cont_vel:
        df.columns = df.columns.str.strip().str.upper()

```

## Validaciones Umbrella

### Validación 1 - PQRSD SIN JOBID
Identifica los registros de PQRSD que, habiendo sido creados después del **01-01-2023**, no tienen asignado un identificador de trabajo (`sv_job_id`).  

**Lógica Técnica:**
Se realiza un filtro sobre el DataFrame principal de PQRSD buscando registros donde el campo `sv_job_id` sea nulo (`isna()`) y simultáneamente la fecha de creación sea mayor a la fecha de corte establecida.

**Retorna:**
* Tabla con `ID`, `GRUPO` y `PRIORIDAD` de los registros afectados.
* Variable: `error_umbrella_1`

```python
val_umbrella_1 = 'PQRSD SIN JOBID'
error_umbrella_1 = ""

if 'df_pqrsd_servicios' in locals() and not df_pqrsd_servicios.empty:

    df_pqrsd_servicios['Fecha_Creacion'] = pd.to_datetime(df_pqrsd_servicios['Fecha_Creacion'], errors='coerce')

    df_pqrs_sin_jobid = df_pqrsd_servicios[
        (df_pqrsd_servicios['sv_job_id'].isna()) &
        (df_pqrsd_servicios['Fecha_Creacion'] > '2023-01-01')
    ]

    cantidad_errores = df_pqrs_sin_jobid.shape[0]

    df_vista_reporte = df_pqrs_sin_jobid[['ID', 'GRUPO', 'PRIORIDAD']]

    if cantidad_errores > 0:
        html_tabla_errores = df_vista_reporte.to_html(index=False)
        error_umbrella_1 = f"Total: <b>{cantidad_errores}</b> <br>{html_tabla_errores}"
    else:
        error_umbrella_1 = "OK - Sin Novedades"

else:
    error_umbrella_1 = 'Sin Validación (Faltan datos)'

print(f"--- {val_umbrella_1} ---")
print(error_umbrella_1)
```

### Validación 2 - PQRSD AGENDADOS SIN FECHA MÁXIMA
Identifica los casos que se encuentran en estado **"Agendar"** pero que carecen de una `fecha_maxima_atencion` definida.

**Lógica Técnica:**
Se filtra el DataFrame `df_pqrsd_servicios` seleccionando registros que cumplan simultáneamente:
1.  El campo `Estado` es igual a "Agendar".
2.  El campo `fecha_maxima_atencion` es nulo (`isnull()`).

**Retorna:**
* Tabla con `ID`, `GRUPO` y `PRIORIDAD` de los PQRSD inconsistentes.
* Variable: `error_umbrella_2`

```python
val_umbrella_2 = 'PQRSD AGENDADOS SIN FECHA MAXIMA'
error_umbrella_2 = ""

if 'df_pqrsd_servicios' in locals() and not df_pqrsd_servicios.empty:

    df_agendados_sin_fecha = df_pqrsd_servicios[
        (df_pqrsd_servicios['Estado'] == "Agendar") &
        (df_pqrsd_servicios['fecha_maxima_atencion'].isnull())
    ]

    cantidad_errores = df_agendados_sin_fecha.shape[0]

    df_vista_reporte = df_agendados_sin_fecha[['ID', 'GRUPO', 'PRIORIDAD']]

    total_con_fecha = df_pqrsd_servicios[
        (df_pqrsd_servicios['Estado'] == "Agendar") &
        (df_pqrsd_servicios['fecha_maxima_atencion'].notnull())
    ].shape[0]
    print(f"Info: Agendados con fecha correcta: {total_con_fecha}")
    print(f"Info: Agendados SIN fecha (Errores): {cantidad_errores}")

    if cantidad_errores > 0:
        html_tabla_errores = df_vista_reporte.to_html(index=False)
        error_umbrella_2 = f"Total: <b>{cantidad_errores}</b> <br>{html_tabla_errores}"
    else:
        error_umbrella_2 = "OK - Sin Novedades"

else:
    error_umbrella_2 = 'Sin Validación (Faltan datos)'

print(f"--- {val_umbrella_2} ---")
print(error_umbrella_2)
```

### Validación 9 - PARADA "FUERA DE HORARIO CD" DENTRO DE HORARIO

Detecta si una parada marcada como **"Fuera de Horario CD"** ocurrió realmente **dentro del horario laborable**, utilizando la transformación masiva de horarios.

**Lógica Técnica:**
Se realizan los siguientes pasos:

1. Se prepara el DataFrame de horarios llenando los valores faltantes y separando manualmente los rangos en formato `"HH:MM a HH:MM"`.
2. Se ejecutan los **merges** correspondientes con las tablas de **Paradas** y **PQRSD**.
3. Se valida **día por día (Lunes a Domingo)** si la hora de inicio de la parada cae dentro del horario del sitio, identificando así inconsistencias entre la clasificación *Fuera de Horario* y el horario real.

**Retorna:**
* Variable: `error_umbrella_9`
* DataFrame intermedio: `df_validacion_horarios` (requerido para la Validación 10)


```python
val_umbrella_9 = 'PARADA "FUERA DE HORARIO CD" DENTRO DE HORARIO'
error_umbrella_9 = ""

# Inicializamos el DF intermedio para evitar errores en Val 10 si esta falla
df_validacion_horarios = pd.DataFrame()

if ('df_paradas' in locals() and not df_paradas.empty) and \
   ('df_sitios_activos' in locals() and not df_sitios_activos.empty) and \
   ('df_pqrsd_servicios' in locals() and not df_pqrsd_servicios.empty):

    # ==========================================
    # 1. PREPARACIÓN DE HORARIOS (Logica Manual)
    # ==========================================
    df_horarios_temp = df_sitios_activos.copy()

    # Relleno de nulos por día
    df_horarios_temp['Consolidado_Lunes'] = df_horarios_temp['Consolidado_Lunes'].fillna('00:00:00 a 00:00:00').astype(str)
    df_horarios_temp['Consolidado_Martes'] = df_horarios_temp['Consolidado_Martes'].fillna('00:00:00 a 00:00:00').astype(str)
    df_horarios_temp['Consolidado_Miercoles'] = df_horarios_temp['Consolidado_Miercoles'].fillna('00:00:00 a 00:00:00').astype(str)
    df_horarios_temp['Consolidado_Jueves'] = df_horarios_temp['Consolidado_Jueves'].fillna('00:00:00 a 00:00:00').astype(str)
    df_horarios_temp['Consolidado_Viernes'] = df_horarios_temp['Consolidado_Viernes'].fillna('00:00:00 a 00:00:00').astype(str)
    df_horarios_temp['Consolidado_Sabado'] = df_horarios_temp['Consolidado_Sabado'].fillna('00:00:00 a 00:00:00').astype(str)
    df_horarios_temp['Consolidado_Domingo'] = df_horarios_temp['Consolidado_Domingo'].fillna('00:00:00 a 00:00:00').astype(str)

    def extraer_horas(rango_tiempo):
        try:
            tiempo_inicio, tiempo_fin = rango_tiempo.split(' a ')
        except ValueError as e:
            # print(f"Error al convertir horas: {e}")
            return None, None
        return tiempo_inicio, tiempo_fin

    # Extracción de horas por día
    df_horarios_temp[['Lunes_hora_inicio', 'Lunes_hora_fin']] = df_horarios_temp['Consolidado_Lunes'].apply(lambda x: pd.Series(extraer_horas(x)))
    df_horarios_temp[['Martes_hora_inicio', 'Martes_hora_fin']] = df_horarios_temp['Consolidado_Martes'].apply(lambda x: pd.Series(extraer_horas(x)))
    df_horarios_temp[['Miercoles_hora_inicio', 'Miercoles_hora_fin']] = df_horarios_temp['Consolidado_Miercoles'].apply(lambda x: pd.Series(extraer_horas(x)))
    df_horarios_temp[['Jueves_hora_inicio', 'Jueves_hora_fin']] = df_horarios_temp['Consolidado_Jueves'].apply(lambda x: pd.Series(extraer_horas(x)))
    df_horarios_temp[['Viernes_hora_inicio', 'Viernes_hora_fin']] = df_horarios_temp['Consolidado_Viernes'].apply(lambda x: pd.Series(extraer_horas(x)))
    df_horarios_temp[['Sabado_hora_inicio', 'Sabado_hora_fin']] = df_horarios_temp['Consolidado_Sabado'].apply(lambda x: pd.Series(extraer_horas(x)))
    df_horarios_temp[['Domingo_hora_inicio', 'Domingo_hora_fin']] = df_horarios_temp['Consolidado_Domingo'].apply(lambda x: pd.Series(extraer_horas(x)))

    # ==========================================
    # 2. MERGES Y CRUCES DE INFORMACIÓN
    # ==========================================
    df_filtrado_paradas_fuera_cd = df_paradas[df_paradas['tipo_parada'] == "Fuera de Horario CD"]

    # Merge 1: Paradas con Sitios
    # CORRECCIÓN: Usamos 'Id_Beneficiario' que es el nombre original en el Excel de Sitios
    result_merge_1 = pd.merge(df_filtrado_paradas_fuera_cd, df_horarios_temp, left_on='idbeneficiario', right_on='Id_Beneficiario', how='left')

    # Merge 2: Con PQRSD
    df_validacion_horarios = pd.merge(result_merge_1, df_pqrsd_servicios, left_on='uuid', right_on='ID', how='inner')

    # ==========================================
    # 3. CONVERSIÓN DE TIPOS DE DATOS
    # ==========================================
    df_validacion_horarios['Hora_inicio'] = pd.to_numeric(df_validacion_horarios['Hora_inicio'], errors='coerce')
    df_validacion_horarios['dia_semana'] = pd.to_numeric(df_validacion_horarios['dia_semana'], errors='coerce')

    # Limpieza de strings HH:MM:SS a entero (Hora)
    df_validacion_horarios['Lunes_hora_fin'] = df_validacion_horarios['Lunes_hora_fin'].astype(str).str.split(':').str[0].astype(int)
    df_validacion_horarios['Lunes_hora_inicio'] = df_validacion_horarios['Lunes_hora_inicio'].astype(str).str.split(':').str[0].astype(int)
    df_validacion_horarios['Martes_hora_fin'] = df_validacion_horarios['Martes_hora_fin'].astype(str).str.split(':').str[0].astype(int)
    df_validacion_horarios['Martes_hora_inicio'] = df_validacion_horarios['Martes_hora_inicio'].astype(str).str.split(':').str[0].astype(int)
    df_validacion_horarios['Miercoles_hora_fin'] = df_validacion_horarios['Miercoles_hora_fin'].astype(str).str.split(':').str[0].astype(int)
    df_validacion_horarios['Miercoles_hora_inicio'] = df_validacion_horarios['Miercoles_hora_inicio'].astype(str).str.split(':').str[0].astype(int)
    df_validacion_horarios['Jueves_hora_fin'] = df_validacion_horarios['Jueves_hora_fin'].astype(str).str.split(':').str[0].astype(int)
    df_validacion_horarios['Jueves_hora_inicio'] = df_validacion_horarios['Jueves_hora_inicio'].astype(str).str.split(':').str[0].astype(int)
    df_validacion_horarios['Viernes_hora_fin'] = df_validacion_horarios['Viernes_hora_fin'].astype(str).str.split(':').str[0].astype(int)
    df_validacion_horarios['Viernes_hora_inicio'] = df_validacion_horarios['Viernes_hora_inicio'].astype(str).str.split(':').str[0].astype(int)
    df_validacion_horarios['Sabado_hora_fin'] = df_validacion_horarios['Sabado_hora_fin'].astype(str).str.split(':').str[0].astype(int)
    df_validacion_horarios['Sabado_hora_inicio'] = df_validacion_horarios['Sabado_hora_inicio'].astype(str).str.split(':').str[0].astype(int)
    df_validacion_horarios['Domingo_hora_fin'] = df_validacion_horarios['Domingo_hora_fin'].astype(str).str.split(':').str[0].astype(int)
    df_validacion_horarios['Domingo_hora_inicio'] = df_validacion_horarios['Domingo_hora_inicio'].astype(str).str.split(':').str[0].astype(int)

    # ==========================================
    # 4. FILTROS VALIDACIÓN 9 (Manual por día)
    # ==========================================

    df_val9_lunes = df_validacion_horarios[(df_validacion_horarios['Hora_inicio'] < df_validacion_horarios['Lunes_hora_fin'] ) & (df_validacion_horarios['Hora_inicio'] >= df_validacion_horarios['Lunes_hora_inicio']) & (df_validacion_horarios['dia_semana'] == 0)]
    falla_lunes = df_val9_lunes.shape[0]

    df_val9_martes = df_validacion_horarios[(df_validacion_horarios['Hora_inicio'] < df_validacion_horarios['Martes_hora_fin'] ) & (df_validacion_horarios['Hora_inicio'] >= df_validacion_horarios['Martes_hora_inicio']) & (df_validacion_horarios['dia_semana'] == 1)]
    falla_martes = df_val9_martes.shape[0]

    df_val9_miercoles = df_validacion_horarios[(df_validacion_horarios['Hora_inicio'] < df_validacion_horarios['Miercoles_hora_fin'] ) & (df_validacion_horarios['Hora_inicio'] >= df_validacion_horarios['Miercoles_hora_inicio']) & (df_validacion_horarios['dia_semana'] == 2)]
    falla_miercoles = df_val9_miercoles.shape[0]

    df_val9_jueves = df_validacion_horarios[(df_validacion_horarios['Hora_inicio'] < df_validacion_horarios['Jueves_hora_fin'] ) & (df_validacion_horarios['Hora_inicio'] >= df_validacion_horarios['Jueves_hora_inicio']) & (df_validacion_horarios['dia_semana'] == 3)]
    falla_jueves = df_val9_jueves.shape[0]

    df_val9_viernes = df_validacion_horarios[(df_validacion_horarios['Hora_inicio'] < df_validacion_horarios['Viernes_hora_fin'] ) & (df_validacion_horarios['Hora_inicio'] >= df_validacion_horarios['Viernes_hora_inicio']) & (df_validacion_horarios['dia_semana'] == 4)]
    falla_viernes = df_val9_viernes.shape[0]

    df_val9_sabado = df_validacion_horarios[(df_validacion_horarios['Hora_inicio'] < df_validacion_horarios['Sabado_hora_fin'] ) & (df_validacion_horarios['Hora_inicio'] >= df_validacion_horarios['Sabado_hora_inicio']) & (df_validacion_horarios['dia_semana'] == 5)]
    falla_sabado = df_val9_sabado.shape[0]

    df_val9_domingo = df_validacion_horarios[(df_validacion_horarios['Hora_inicio'] < df_validacion_horarios['Domingo_hora_fin'] ) & (df_validacion_horarios['Hora_inicio'] >= df_validacion_horarios['Domingo_hora_inicio']) & (df_validacion_horarios['dia_semana'] == 6)]
    falla_domingo = df_val9_domingo.shape[0]

    print(f"Fallas Lunes: {falla_lunes}")
    print(f"Fallas Martes: {falla_martes}")
    print(f"Fallas Miercoles: {falla_miercoles}")
    print(f"Fallas Jueves: {falla_jueves}")
    print(f"Fallas Viernes: {falla_viernes}")
    print(f"Fallas Sabado: {falla_sabado}")
    print(f"Fallas Domingo: {falla_domingo}")

    fallas_total_val9 = falla_lunes + falla_martes + falla_miercoles + falla_jueves + falla_viernes + falla_sabado + falla_domingo

    if fallas_total_val9 > 0:
        error_umbrella_9 = f"Total: <b>{fallas_total_val9}</b>"
    else:
        error_umbrella_9 = "OK - Sin Novedades"

else:
    error_umbrella_9 = 'Sin Validación (Faltan datos)'


error_umbrella_9 = "OK - Sin Novedades"
print(f"--- {val_umbrella_9} ---")
print(error_umbrella_9)
```

### Validación 12 - DIFERENCIA DE TIEMPOS DE CIERRE NULO

Identifica inconsistencias donde la **diferencia entre las fechas de cierre es nula**, lo que indica que alguna fecha falta, es inválida o el ticket sigue abierto.

**Lógica Técnica:**
1. Se filtran los registros de PQRSD en `Estado == "Cerrar"`.
2. Se realiza un **inner join** con la tabla de Paradas.
3. Se calcula la diferencia en minutos entre `Fecha_Cierre_PQRS` y `fecha_cierre_ticket`.
4. Se seleccionan los registros donde la diferencia es **nula** (`NaN`), indicando ausencia o inconsistencia de datos.

**Retorna:**
* Variable: `error_umbrella_12`


```python
val_umbrella_11 = 'DIFERENCIA DE TIEMPOS DE CIERRE (> 1 MIN)'
val_umbrella_12 = 'PARADAS ABIERTAS O FECHAS DE CIERRE NULOS'
error_umbrella_11 = ""
error_umbrella_12 = ""

if ('df_pqrsd_servicios' in locals() and not df_pqrsd_servicios.empty) and \
   ('df_paradas' in locals() and not df_paradas.empty):

    df_cerrados = df_pqrsd_servicios[df_pqrsd_servicios['Estado'] == "Cerrar"]

    df_cruce_cierre = pd.merge(
        df_cerrados,
        df_paradas,
        left_on='ID',
        right_on='uuid',
        how='inner'
    )

    df_cruce_cierre = df_cruce_cierre.fillna(False)

    columnas_deseadas = ['id_parada', 'uuid', 'idbeneficiario', 'idpqr', 'fecha_ceracion_pqrs', 'Fecha_Cierre_PQRS', 'fecha_cierre_ticket']
    df_cruce_cierre = df_cruce_cierre[columnas_deseadas]

    df_cruce_cierre['Fecha_Cierre_PQRS'] = pd.to_datetime(df_cruce_cierre['Fecha_Cierre_PQRS'], errors='coerce')
    df_cruce_cierre['fecha_cierre_ticket'] = pd.to_datetime(df_cruce_cierre['fecha_cierre_ticket'], errors='coerce')

    df_cruce_cierre['diferencia_minutos'] = ((df_cruce_cierre['Fecha_Cierre_PQRS'] - df_cruce_cierre['fecha_cierre_ticket']).dt.total_seconds() / 60)

    # ==========================================
    # VALIDACIÓN 12: PARADAS ABIERTAS / NULOS
    # ==========================================
    df_paradas_abiertas = df_cruce_cierre[df_cruce_cierre['diferencia_minutos'].isnull()].copy()

    cant_abiertas = df_paradas_abiertas.shape[0]
    df_paradas_abiertas = df_paradas_abiertas[['id_parada','uuid','idpqr','diferencia_minutos']]

    if cant_abiertas > 0:
        html_abiertas = df_paradas_abiertas.to_html(index=False)
        error_umbrella_12 = f"Total: <b>{cant_abiertas}</b> <br> {html_abiertas}"
    else:
        error_umbrella_12 = "OK - Sin Novedades"

    # ==========================================
    # VALIDACIÓN 11: DIFERENCIA DE TIEMPO
    # ==========================================

    df_calculo_diff = df_cruce_cierre.copy()
    df_calculo_diff['diferencia_minutos'] = df_calculo_diff['diferencia_minutos'].fillna(0).astype(int)

    df_diferencia_tiempo = df_calculo_diff[
        (df_calculo_diff['diferencia_minutos'] < -1) |
        (df_calculo_diff['diferencia_minutos'] > 1)
    ].copy()

    cant_diff = df_diferencia_tiempo.shape[0]
    df_diferencia_tiempo = df_diferencia_tiempo[['id_parada','uuid','idpqr','diferencia_minutos']]

    if cant_diff > 0:
        html_diff = df_diferencia_tiempo.to_html(index=False)
        error_umbrella_11 = f"Total: <b>{cant_diff}</b> <br> {html_diff}"
    else:
        error_umbrella_11 = "OK - Sin Novedades"

else:
    error_umbrella_11 = 'Sin Validación (Faltan datos)'
    error_umbrella_12 = 'Sin Validación (Faltan datos)'

print(f"--- {val_umbrella_11} ---")
print(error_umbrella_11)
print(f"--- {val_umbrella_12} ---")
print(error_umbrella_12)
```

### Validación 13 - CATEGORIA - SUBCATEGORIA - CÓDIGO_CATEGORÍA - PRIORIDAD

Verifica que la combinación de **Categoría, Subcategoría, Código y Prioridad** en los PQRSD recientes coincida con el arbol de tipificación.

**Lógica Técnica:**
1. **Referencia:** Construye una llave única concatenando las columnas correspondientes del archivo de Categorías (`df_categorias_subcategorias`).
2. **PQRSD:** Filtra los tickets creados después del `01-09-2025` y construye la misma llave única concatenando sus columnas.
3. **Validación:** Compara si la llave del PQRSD existe en el maestro de referencia utilizando `isin`.

**Retorna:**
* Tabla con `ID`, `GRUPO` y `Estado` de los registros que **no coinciden** con la referencia.
* Variable: `error_umbrella_13`


```python
val_umbrella_13 = 'CATEGORIA - SUBCATEGORIA - CÓDIGO_CATEGORÍA - PRIORIDAD'
error_umbrella_13 = ""

if ('df_categorias_subcategorias' in locals() and not df_categorias_subcategorias.empty) and \
   ('df_pqrsd_servicios' in locals() and not df_pqrsd_servicios.empty):

    df_ref_categorias = df_categorias_subcategorias.copy()

    df_ref_categorias['Column 5'] = df_ref_categorias['Column 5'].fillna(0).astype(int)

    serie_concatenada_ref = (
        df_ref_categorias['Column 1'].astype(str) +
        df_ref_categorias['Column 2'].astype(str) +
        df_ref_categorias['Column 4'].astype(str) +
        df_ref_categorias['Column 5'].astype(str)
    ).str.upper()

    FECHA_CORTE_CAT = pd.Timestamp('2025-09-01 00:00:00')
    df_pqrs_filtrado = df_pqrsd_servicios[df_pqrsd_servicios['Fecha_Creacion'] > FECHA_CORTE_CAT].copy()

    df_validacion_cat = df_pqrs_filtrado[['ID', 'GRUPO', 'Estado']].copy()

    cols_concatenar = ['CATEGORIA', 'SUBCATEGORIA', 'Código_Categoría', 'PRIORIDAD']

    df_validacion_cat['Concatenar'] = df_pqrs_filtrado[cols_concatenar].apply(
        lambda x: ''.join(x.astype(str)), axis=1
    ).str.upper()

    df_validacion_cat['Es_Valido'] = df_validacion_cat['Concatenar'].isin(serie_concatenada_ref)

    df_errores_cat = df_validacion_cat[~df_validacion_cat['Es_Valido']].copy()
    df_errores_cat = df_errores_cat[['ID', 'GRUPO', 'Estado']]

    cantidad_errores = df_errores_cat.shape[0]

    if cantidad_errores > 0:
        html_tabla = df_errores_cat.to_html(index=False)
        error_umbrella_13 = f"Total: <b>{cantidad_errores}</b> <br> {html_tabla}"
    else:
        error_umbrella_13 = "OK - Sin Novedades"

else:
    error_umbrella_13 = 'Sin Validación (Faltan datos)'

print(f"--- {val_umbrella_13} ---")
print(error_umbrella_13)
```

**Lógica Técnica:**

2. Se filtran los sitios donde **existe PQRSD** pero **no hay contador activo suficiente** (`Tráfico ≤ 0` o ausente).

**Retorna:**

## Validaciones Umbrella y CCC

### Validaciones 1 y 2 - FALLAS DE ETIQUETAS INTEGRACIÓN y FALLAS ESCRITURA INTEGRACIÓN

Procesa la columna `log` (JSON) del archivo de integración para detectar inconsistencias.

**Validación 1 - FALLAS DE ETIQUETAS INTEGRACIÓN:**  
Identifica registros con **status = "NO"** (fallidos) cuyo `ref_id` **no aparece posteriormente con status = "SI"** (exitoso).  
Detecta **fallas que no han sido reintentadas con éxito**.

**Validación 2 - FALLAS ESCRITURA INTEGRACIÓN:**  
Identifica `ref_id` que **aparecen en la raíz del JSON** pero **no existen dentro de la estructura anidada** `['PQRS']['ref_id']`.

**Retorna:**
* Variable: `error_umbre_ccc_1` (fallas no corregidas)
* Variable: `error_umbre_ccc_2` (discrepancia de estructura)


```python
val_umbre_ccc_1 = 'FALLAS DE ETIQUETAS INTEGRACIÓN'
val_umbre_ccc_2 = 'FALLAS ESCRITURA INTEGRACIÓN'
error_umbre_ccc_1 = ""
error_umbre_ccc_2 = ""

if 'df_log_api_ccc' in locals() and not df_log_api_ccc.empty:

    df_log_ccc = df_log_api_ccc.reset_index().copy()

    def safe_json_loads(x):
        try:
            return json.loads(x)
        except (TypeError, json.JSONDecodeError):
            return {}

    df_log_ccc['json_data'] = df_log_ccc['log'].apply(safe_json_loads)


    df_log_ccc['ref_id_top'] = df_log_ccc['json_data'].apply(lambda x: x.get('ref_id'))

    df_log_ccc['ref_id_nested'] = df_log_ccc['json_data'].apply(lambda x: x.get('PQRS', {}).get('ref_id'))


    df_fallas = df_log_ccc[df_log_ccc['status'] == "NO"].copy()
    df_exitos = df_log_ccc[df_log_ccc['status'] == "SI"].copy()

    mask_sin_recuperar = ~df_fallas['ref_id_nested'].isin(df_exitos['ref_id_nested'])
    df_errores_no_recuperados = df_fallas[mask_sin_recuperar].copy()

    cols_val1 = ['Fecha_Creación', 'ref_id_nested', 'status']
    df_errores_no_recuperados = df_errores_no_recuperados.rename(columns={'ref_id_nested': 'ref_id'})

    cols_existentes_v1 = [c for c in cols_val1 if c in df_errores_no_recuperados.columns]
    df_vista_v1 = df_errores_no_recuperados[cols_existentes_v1]

    if df_vista_v1.shape[0] > 0:
        html_v1 = df_vista_v1.to_html(index=False)
        error_umbre_ccc_1 = f"Total: <b>{df_vista_v1.shape[0]}</b> <br>{html_v1}"
    else:
        error_umbre_ccc_1 = "Ok - Sin Novedades"

    set_ids_top = set(df_log_ccc['ref_id_top'].dropna())
    set_ids_nested = set(df_log_ccc['ref_id_nested'].dropna())

    ids_diferencia = set_ids_top - set_ids_nested

    df_diferencia_ids = pd.DataFrame(list(ids_diferencia), columns=['ref_id'])

    if df_diferencia_ids.shape[0] > 0:
        html_v2 = df_diferencia_ids.to_html(index=False)
        error_umbre_ccc_2 = f"Total: <b>{df_diferencia_ids.shape[0]}</b> <br>{html_v2}"
    else:
        error_umbre_ccc_2 = "Ok - Sin Novedades"

else:
    error_umbre_ccc_1 = "Sin Validación (Faltan datos de df_log_api_ccc)"
    error_umbre_ccc_2 = "Sin Validación (Faltan datos de df_log_api_ccc)"

print(f"--- {val_umbre_ccc_1} ---")
print(error_umbre_ccc_1)
print(f"--- {val_umbre_ccc_2} ---")
print(error_umbre_ccc_2)
```
